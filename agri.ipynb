```python
for col in df.columns:
    if "AREA" in col and "1000 ha" in col:
        df[col] = df[col] * 1000
        df.rename(columns={col: col.replace("1000 ha", "HA")}, inplace=True)
```

```python
display(df)
```

```python

for col in df.columns:
    if "PRODUCTION" in col and "1000 tons" in col:
        df[col] = df[col] * 1_000_000
        df.rename(columns={col: col.replace("1000 tons", "KG")}, inplace=True)
```

```python
display(df)
```

```python
df["State Name"] = df["State Name"].astype(str)
df["Dist Name"] = df["Dist Name"].astype(str)
df["Year"] = df["Year"].astype(int)
```

```python
df.columns = (
    df.columns.str.strip()
    .str.upper()
    .str.replace(" ", "_")
    .str.replace(r"[()]", "", regex=True)
    .str.replace("/", "_")
)
```

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS agriculture_data(
    YEAR INT,
    STATE_NAME VARCHAR(100),
    DIST_NAME VARCHAR(100),
    RICE_AREA DOUBLE,
    RICE_PRODUCTION DOUBLE,
    RICE_YIELD DOUBLE,
    WHEAT_AREA DOUBLE,
    WHEAT_PRODUCTION DOUBLE,
    WHEAT_YIELD DOUBLE,
    KHARIF_SORGHUM_AREA DOUBLE,
    KHARIF_SORGHUM_PRODUCTION DOUBLE,
    KHARIF_SORGHUM_YIELD DOUBLE,
    RABI_SORGHUM_AREA DOUBLE,
    RABI_SORGHUM_PRODUCTION DOUBLE,
    RABI_SORGHUM_YIELD DOUBLE,
    SORGHUM_AREA DOUBLE,
    SORGHUM_PRODUCTION DOUBLE,
    SORGHUM_YIELD DOUBLE,
    PEARL_MILLET_AREA DOUBLE,
    PEARL_MILLET_PRODUCTION DOUBLE,
    PEARL_MILLET_YIELD DOUBLE,
    MAIZE_AREA DOUBLE,
    MAIZE_PRODUCTION DOUBLE,
    MAIZE_YIELD DOUBLE,
    FINGER_MILLET_AREA DOUBLE,
    FINGER_MILLET_PRODUCTION DOUBLE,
    FINGER_MILLET_YIELD DOUBLE,
    GROUNDNUT_AREA DOUBLE,
    GROUNDNUT_PRODUCTION DOUBLE,
    GROUNDNUT_YIELD DOUBLE,
    SUNFLOWER_AREA DOUBLE,
    SUNFLOWER_PRODUCTION DOUBLE,
    SUNFLOWER_YIELD DOUBLE,
    SOYABEAN_AREA DOUBLE,
    SOYABEAN_PRODUCTION DOUBLE,
    SOYABEAN_YIELD DOUBLE,
    OILSEEDS_AREA DOUBLE,
    OILSEEDS_PRODUCTION DOUBLE,
    OILSEEDS_YIELD DOUBLE,
    SUGARCANE_AREA DOUBLE,
    SUGARCANE_PRODUCTION DOUBLE,
    SUGARCANE_YIELD DOUBLE,
    COTTON_AREA DOUBLE,
    COTTON_PRODUCTION DOUBLE,
    COTTON_YIELD DOUBLE
        
);
"""

mycursor.execute("USE agridata;") 
mycursor.execute(create_table_query)
connector.connect()
print("Table 'agriculture_data' created successfully!")
```

```python
try:
    # Convert DataFrame to list of tuples
    data_list = df.values.tolist()
    
    # Connect to the specific database
    connector = pymysql.connect(
        host="localhost",
        user="root",
        password="1234",
        database="agridata"
    )
    
    mycursor = connector.cursor()
    
    query = """
    INSERT INTO agriculture_data (
        YEAR,STATE_NAME,DIST_NAME, 
        RICE_AREA,RICE_PRODUCTION,RICE_YIELD, 
        WHEAT_AREA,WHEAT_PRODUCTION,WHEAT_YIELD,
        KHARIF_SORGHUM_AREA,KHARIF_SORGHUM_PRODUCTION,KHARIF_SORGHUM_YIELD,
        RABI_SORGHUM_AREA,RABI_SORGHUM_PRODUCTION,RABI_SORGHUM_YIELD,
        SORGHUM_AREA,SORGHUM_PRODUCTION,SORGHUM_YIELD,
        PEARL_MILLET_AREA,PEARL_MILLET_PRODUCTION,PEARL_MILLET_YIELD,
        MAIZE_AREA,MAIZE_PRODUCTION,MAIZE_YIELD,
        FINGER_MILLET_AREA,FINGER_MILLET_PRODUCTION,FINGER_MILLET_YIELD,
        GROUNDNUT_AREA,GROUNDNUT_PRODUCTION,GROUNDNUT_YIELD,
        SUNFLOWER_AREA,SUNFLOWER_PRODUCTION,SUNFLOWER_YIELD,
        SOYABEAN_AREA,SOYABEAN_PRODUCTION,SOYABEAN_YIELD,
        OILSEEDS_AREA,OILSEEDS_PRODUCTION,OILSEEDS_YIELD,
        SUGARCANE_AREA,SUGARCANE_PRODUCTION,SUGARCANE_YIELD,
        COTTON_AREA,COTTON_PRODUCTION,COTTON_YIELD
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,
                %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,
                %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
                )
    """
    
    # Insert in batches to prevent memory issues
    batch_size = 1000
    for i in range(0, len(data_list), batch_size):
        batch = data_list[i:i + batch_size]
        mycursor.executemany(query, batch)
        connector.commit()
        print(f"Inserted batch {i//batch_size + 1}, {len(batch)} rows")

    print(f"Total {len(data_list)} rows inserted successfully!")

except pymysql.Error as e:
    print(f"MySQL Error: {e}")
    if connector:
        connector.rollback()
finally:
    if mycursor:
        mycursor.close()
    if connector:
        connector.close()
```
